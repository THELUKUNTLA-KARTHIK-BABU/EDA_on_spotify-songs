```python
# Create a bar graph of the # of tracks in 2023. Sort them from highest to lowest by counts

x = df_2023.groupby('released_month')['released_month'].count().sort_values(ascending=False)

ax = x.plot(kind='bar', rot=0, figsize=(10,5), grid=False, color='purple')
ax.bar_label(ax.containers[0], label_type='center', color='white', weight='bold')
plt.title('Count of Tracks in 2023 by Month')
plt.xlabel('Number of Tracks')
plt.ylabel('Months')
plt.show()
```

```python
# Let's create a dataframe of all tracks from 2021 and 2023

# Dataframe for any track that falls between 2021 and 2023
df_21_23 = df[df['released_year'].isin([2021, 2022, 2023])]
df_21_23.reset_index(inplace=True,drop=True)
```

```python
# List the top 5 rows from 2021 through 2023
df_21_23.head(5)
```

```python


#VISUALIZE DATA FROM 2021-2023

```

```python
# Plot the graph of count of tracks between 2021 and 2023 by Month. Sort them by month

year_mon_pair = ['released_year','released_month']

x = df_21_23.groupby(year_mon_pair)[year_mon_pair].count()

ax = x.plot(kind='bar', rot=90, figsize=(15,5), grid=False, color='purple')
ax.bar_label(ax.containers[0], padding=1, label_type='edge', 
             color='black',  rotation=0)
plt.title('Count of Tracks between 2021 and 2023 by Month')
plt.xlabel('Year and Months in (yyyy, mm)')
plt.ylabel('Number of Tracks')
plt.legend('',frameon=False)
plt.show()
```

```python
# Let's find out the top 15 artists with their tracks in the dataframe

artist_counts = df_21_23.groupby('artist_name')['artist_name'] \
                        .count().sort_values(ascending=False)
artist_counts.head(15)
```
